```python
import pandas as pd

fos19=pd.read_csv('FieldOfStudyData1718_1819_PP.csv')
fos20=pd.read_csv('FieldOfStudyData1819_1920_PP.csv')
fos21=pd.read_csv('FieldOfStudyData1920_2021_PP.csv')

print(fos19.shape)
```

```python
fos_column_names = pd.DataFrame(fos21.columns, columns=['Column Names'])
fos_column_names.to_csv('fos21_column_names.csv', index=False)
```

This will take the data which is every CIP*Year*institution format and create a STEM indicator using NSF classificiations.


```python
from decimal import Decimal, getcontext

nsf=pd.read_csv('NSF_CIP_codes.csv')

#create a new variable that will allow for a recode
nsf['CIP2']=nsf['CIP Code'].apply(lambda x: str(Decimal(str(x)).quantize(Decimal('0.00')))).str.replace(r'\.', '', regex=True)
#fos21['CIPCODE2']=fos21['CIPCODE']
# Now that 'CIP2' is guaranteed to be string type, .str accessor will work
#nsf['CIP2'] = nsf['CIP2'].str.replace('.', '', regex=False)
#print(nsf['CIP Code'].head())
#print(nsf['CIP2'].head())
# Convert to string and remove trailing .0
#fos21['CIPCODE2'] = fos21['CIPCODE2'].replace('.0', '', regex=False)
#print(fos19['CIPCODE'].head())
#print(fos20['CIPCODE'].head())
#print(fos21['CIPCODE'].head())



#now create indicators for each year of the fos to designate NSF STEM classifications by CIPCODE and CIP2
fos19['STEM_CIP'] = fos19['CIPCODE'].apply(lambda x: x if str(x) in nsf['CIP2'].values else 'Non-STEM') # Convert x to string for consistent comparison
fos19['STEM'] = fos19['CIPCODE'].apply(lambda x: 1 if str(x) in nsf['CIP2'].values else 0) # Convert x to string for consistent comparison

fos20['STEM_CIP'] = fos20['CIPCODE'].apply(lambda x: x if str(x) in nsf['CIP2'].values else 'Non-STEM') # Convert x to string for consistent comparison
fos20['STEM'] = fos20['CIPCODE'].apply(lambda x: 1 if str(x) in nsf['CIP2'].values else 0) # Convert x to string for consistent comparison

fos21['STEM_CIP'] = fos21['CIPCODE'].apply(lambda x: x if str(x) in nsf['CIP2'].values else 'Non-STEM') # Convert x to string for consistent comparison
fos21['STEM'] = fos21['CIPCODE'].apply(lambda x: 1 if str(x) in nsf['CIP2'].values else 0) # Convert x to string for consistent comparison

#check the counts-pretty close each year
print(fos19['STEM'].value_counts())
print(fos20['STEM'].value_counts())
print(fos21['STEM'].value_counts())
```

This tabulates the two new variables STEM_CIP (non-STEM or the CIP code) and STEM (0 means not STEM CIP and 1 means STEM CIP)

```python
print(fos21['STEM_CIP'].value_counts())
print(fos21['STEM'].value_counts())
```

This creates a subset of the data using just the needed columns

```python
fos19_subset = fos19[['UNITID', 'CIPCODE','STEM', 'CREDDESC','IPEDSCOUNT1','IPEDSCOUNT2']].copy()
display(fos19_subset.head())

fos20_subset = fos20[['UNITID', 'CIPCODE','STEM', 'CREDDESC','IPEDSCOUNT1','IPEDSCOUNT2']].copy()
display(fos20_subset.head())

fos21_subset = fos21[['UNITID', 'CIPCODE','STEM', 'CREDDESC','IPEDSCOUNT1','IPEDSCOUNT2']].copy()
display(fos21_subset.head())
```

```python
#replace NaN in IPEDSCOUNT vars with 0
fos19_subset['IPEDSCOUNT1'] = fos19_subset['IPEDSCOUNT1'].fillna(0)
fos19_subset['IPEDSCOUNT2'] = fos19_subset['IPEDSCOUNT2'].fillna(0)

fos20_subset['IPEDSCOUNT1'] = fos20_subset['IPEDSCOUNT1'].fillna(0)
fos20_subset['IPEDSCOUNT2'] = fos20_subset['IPEDSCOUNT2'].fillna(0)

fos21_subset['IPEDSCOUNT1'] = fos21_subset['IPEDSCOUNT1'].fillna(0)
fos21_subset['IPEDSCOUNT2'] = fos21_subset['IPEDSCOUNT2'].fillna(0)

display(fos21_subset.head())
fos21_subset.shape
```

Now widen the data so that each row is a unique UNITID*CREDDESC*STEM classification

```python
fos19_wide = fos19_subset.pivot_table(index=['UNITID', 'CREDDESC','STEM'], values=['IPEDSCOUNT1', 'IPEDSCOUNT2'], aggfunc='sum').reset_index()
display(fos19_wide.head())

fos20_wide = fos20_subset.pivot_table(index=['UNITID', 'CREDDESC','STEM'], values=['IPEDSCOUNT1', 'IPEDSCOUNT2'], aggfunc='sum').reset_index()
display(fos20_wide.head())

fos21_wide = fos21_subset.pivot_table(index=['UNITID', 'CREDDESC','STEM'], values=['IPEDSCOUNT1', 'IPEDSCOUNT2'], aggfunc='sum').reset_index()
display(fos21_wide.head())
```

```python
fos19_long=fos19_wide.melt(id_vars=['UNITID', 'CREDDESC','STEM'], var_name='IPEDSCOUNT', value_name='COUNT')
fos20_long=fos20_wide.melt(id_vars=['UNITID', 'CREDDESC','STEM'], var_name='IPEDSCOUNT', value_name='COUNT')
fos21_long=fos21_wide.melt(id_vars=['UNITID', 'CREDDESC','STEM'], var_name='IPEDSCOUNT', value_name='COUNT')
display(fos19_long.head())
```

This creates a long version of the fos data for merging

```python
fos19_long['Year'] = 2019
fos20_long['Year'] = 2020
fos21_long['Year'] = 2021

combined_fos_long = pd.concat([fos19_long, fos20_long, fos21_long], ignore_index=True)

display(combined_fos_long.head())
display(combined_fos_long.tail())

combined_fos_long.to_csv('combined_fos_long.csv', index=False)
```

This creates a wide version of the fos data for merging.....is this the simplest way????

```python
merged_19_20=pd.merge(fos19_wide,fos20_wide, on=['UNITID', 'CREDDESC', 'STEM'], suffixes=('_2019', '_2020'))

merged_all_years = pd.merge(merged_19_20, fos21_wide, on=['UNITID', 'CREDDESC', 'STEM'], suffixes=('', '_2021'))

# Rename columns for clarity
merged_all_years.rename(columns={
    'IPEDSCOUNT1_2019': 'IPEDSCOUNT1_1718',
    'IPEDSCOUNT2_2019': 'IPEDSCOUNT2_1819',
    'IPEDSCOUNT1_2020': 'IPEDSCOUNT1_1819',
    'IPEDSCOUNT2_2020': 'IPEDSCOUNT2_1920',
    'IPEDSCOUNT1': 'IPEDSCOUNT1_1920',
    'IPEDSCOUNT2': 'IPEDSCOUNT2_2021'
}, inplace=True)

display(merged_all_years.head())

#write to csv
merged_all_years.to_csv('fos_wide_merged.csv', index=False)
```
